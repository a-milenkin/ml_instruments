–¢–∞–∫ –∂–µ –≤ —Ç–µ–∫—Å—Ç–æ–≤–æ–º –ø–æ–ª–µ –ø—Ä–∏–ª–æ–∂–∏—Ç–µ —Å—Å—ã–ª–∫—É —Å —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞–º–∏, –∫–æ—Ç–æ—Ä—É—é –º–æ–∂–Ω–æ —Å–æ–∑–¥–∞—Ç—å –≤ `ClearML` –Ω–∞ —Å—Ç—Ä–∞–Ω–∏—Ü–µ —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–∞ –≤ `–º–µ–Ω—é -> Share`.

```python
import os
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader
import torchvision
import torchvision.datasets as datasets
import torchvision.transforms as transforms
```

```python
# –ì–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä—ã
batch_size = 64
lr = 0.0002
num_epochs = 10
noise_dim = 100
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
# –ü–æ–¥–≥–æ—Ç–æ–≤–∫–∞ –¥–∞—Ç–∞—Å–µ—Ç–∞ MNIST
transform = transforms.Compose(
    [transforms.ToTensor(), transforms.Normalize((0.5,), (0.5,))]
)
dataset = datasets.MNIST(root="./data", train=True, download=True, transform=transform)
dataloader = DataLoader(dataset, batch_size=batch_size, shuffle=True)
```

```python
# –û–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –≥–µ–Ω–µ—Ä–∞—Ç–æ—Ä–∞
class Generator(nn.Module):
    def __init__(self, noise_dim):
        super(Generator, self).__init__()
        self.main = nn.Sequential(
            # –í—Ö–æ–¥: –≤–µ–∫—Ç–æ—Ä —à—É–º–∞ —Ä–∞–∑–º–µ—Ä–∞ noise_dim
            nn.Linear(noise_dim, 256 * 7 * 7),
            nn.ReLU(True),
            nn.Unflatten(1, (256, 7, 7)),
            # –°–æ—Å—Ç–æ—è–Ω–∏–µ: (256, 7, 7)
            nn.ConvTranspose2d(
                256, 128, kernel_size=4, stride=2, padding=1, bias=False
            ),  # -> (128, 14, 14)
            nn.BatchNorm2d(128),
            nn.ReLU(True),
            nn.ConvTranspose2d(
                128, 1, kernel_size=4, stride=2, padding=1, bias=False
            ),  # -> (1, 28, 28)
            nn.Tanh(),
        )

    def forward(self, input):
        return self.main(input)


# –û–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –¥–∏—Å–∫—Ä–∏–º–∏–Ω–∞—Ç–æ—Ä–∞
class Discriminator(nn.Module):
    def __init__(self):
        super(Discriminator, self).__init__()
        self.main = nn.Sequential(
            # –í—Ö–æ–¥: –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–µ (1, 28, 28)
            nn.Conv2d(
                1, 64, kernel_size=4, stride=2, padding=1, bias=False
            ),  # -> (64, 14, 14)
            nn.LeakyReLU(0.2, inplace=True),
            nn.Conv2d(
                64, 128, kernel_size=4, stride=2, padding=1, bias=False
            ),  # -> (128, 7, 7)
            nn.BatchNorm2d(128),
            nn.LeakyReLU(0.2, inplace=True),
            nn.Flatten(),
            nn.Linear(128 * 7 * 7, 1),
            nn.Sigmoid(),
        )

    def forward(self, input):
        return self.main(input)
```

```python
# –°–æ–∑–¥–∞–µ–º –º–æ–¥–µ–ª–∏
netG = Generator(noise_dim).to(device)
netD = Discriminator().to(device)

# –û–ø—Ä–µ–¥–µ–ª—è–µ–º —Ñ—É–Ω–∫—Ü–∏—é –ø–æ—Ç–µ—Ä—å –∏ –æ–ø—Ç–∏–º–∏–∑–∞—Ç–æ—Ä—ã
criterion = nn.BCELoss()
optimizerD = optim.Adam(netD.parameters(), lr=lr, betas=(0.5, 0.999))
optimizerG = optim.Adam(netG.parameters(), lr=lr, betas=(0.5, 0.999))

# –ú–µ—Ç–∫–∏
real_label = 1.0
fake_label = 0.0
```

```python
# –û—Å–Ω–æ–≤–Ω–æ–π —Ü–∏–∫–ª –æ–±—É—á–µ–Ω–∏—è
for epoch in range(num_epochs):
    for i, (data, _) in enumerate(dataloader):
        ############################
        #  –û–±–Ω–æ–≤–ª—è–µ–º –¥–∏—Å–∫—Ä–∏–º–∏–Ω–∞—Ç–æ—Ä
        ############################
        netD.zero_grad()
        # –û–±—É—á–µ–Ω–∏–µ –Ω–∞ —Ä–µ–∞–ª—å–Ω—ã—Ö –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è—Ö
        real_images = data.to(device)
        b_size = real_images.size(0)
        label = torch.full((b_size,), real_label, dtype=torch.float, device=device)

        output = netD(real_images).view(-1)
        errD_real = criterion(output, label)
        errD_real.backward()
        D_x = output.mean().item()

        # –û–±—É—á–µ–Ω–∏–µ –Ω–∞ —Ñ–µ–π–∫–æ–≤—ã—Ö –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è—Ö
        noise = torch.randn(b_size, noise_dim, device=device)
        fake_images = netG(noise)
        label.fill_(fake_label)
        output = netD(fake_images.detach()).view(-1)
        errD_fake = criterion(output, label)
        errD_fake.backward()
        D_G_z1 = output.mean().item()

        errD = errD_real + errD_fake
        optimizerD.step()

        ############################
        #  –û–±–Ω–æ–≤–ª—è–µ–º –≥–µ–Ω–µ—Ä–∞—Ç–æ—Ä
        ############################
        netG.zero_grad()
        label.fill_(
            real_label
        )  # –î–ª—è –≥–µ–Ω–µ—Ä–∞—Ç–æ—Ä–∞ "—Ñ–µ–π–∫–æ–≤—ã–µ" –º–µ—Ç–∫–∏ —Å—á–∏—Ç–∞—é—Ç—Å—è –∫–∞–∫ —Ä–µ–∞–ª—å–Ω—ã–µ
        output = netD(fake_images).view(-1)
        errG = criterion(output, label)
        errG.backward()
        D_G_z2 = output.mean().item()
        optimizerG.step()

        if i % 300 == 0:
            print(
                "[%d/%d][%d/%d]\tLoss_D: %.4f\tLoss_G: %.4f\tD(x): %.4f\tD(G(z)): %.4f / %.4f"
                % (
                    epoch,
                    num_epochs,
                    i,
                    len(dataloader),
                    errD.item(),
                    errG.item(),
                    D_x,
                    D_G_z1,
                    D_G_z2,
                )
            )

    # –°–æ—Ö—Ä–∞–Ω—è–µ–º —Å—ç–º–ø–ª—ã –≥–µ–Ω–µ—Ä–∞—Ç–æ—Ä–∞ –∫–∞–∂–¥—ã–µ 2 —ç–ø–æ—Ö–∏
    if epoch % 2 == 0:
        with torch.no_grad():
            fixed_noise = torch.randn(64, noise_dim, device=device)
            fake = netG(fixed_noise).detach().cpu()
            os.makedirs("output", exist_ok=True)
            torchvision.utils.save_image(
                fake, f"output/fake_samples_epoch_{epoch}.png", normalize=True
            )
```

```python
%%writefile gan_module.py

import torch
import torch.nn as nn
import torchvision.utils as vutils
import torchvision.transforms as transforms
from lightning import LightningModule
from torch.optim import Adam
from clearml import Task


class Generator(nn.Module):
    def __init__(self, noise_dim):
        super(Generator, self).__init__()
        self.noise_dim = noise_dim
        self.main = nn.Sequential(
            nn.Linear(noise_dim, 256 * 7 * 7),
            nn.ReLU(True),
            nn.Unflatten(1, (256, 7, 7)),
            nn.ConvTranspose2d(256, 128, kernel_size=4, stride=2, padding=1, bias=False),  # -> (128, 14, 14)
            nn.BatchNorm2d(128),
            nn.ReLU(True),
            nn.ConvTranspose2d(128, 1, kernel_size=4, stride=2, padding=1, bias=False),    # -> (1, 28, 28)
            nn.Tanh()
        )
        
    def forward(self, x):
        return self.main(x)

class Discriminator(nn.Module):
    def __init__(self):
        super(Discriminator, self).__init__()
        self.main = nn.Sequential(
            nn.Conv2d(1, 64, kernel_size=4, stride=2, padding=1, bias=False),  # -> (64, 14, 14)
            nn.LeakyReLU(0.2, inplace=True),
            nn.Conv2d(64, 128, kernel_size=4, stride=2, padding=1, bias=False), # -> (128, 7, 7)
            nn.BatchNorm2d(128),
            nn.LeakyReLU(0.2, inplace=True),
            nn.Flatten(),
            nn.Linear(128 * 7 * 7, 1),
            nn.Sigmoid()
        )
        
    def forward(self, x):
        return self.main(x)

class GAN(LightningModule):
    def __init__(self, noise_dim=100, lr=0.0002, betas=(0.5, 0.999), debug_samples_epoch=1):
        super(GAN, self).__init__()
        self.save_hyperparameters()
        self.noise_dim = noise_dim
        self.lr = lr
        self.betas = betas
        self.debug_samples_epoch = debug_samples_epoch
        self.generator = Generator(noise_dim)
        self.discriminator = Discriminator()
        self.criterion = nn.BCELoss()
        # –û—Ç–∫–ª—é—á–∞–µ–º –∞–≤—Ç–æ–º–∞—Ç–∏—á–µ—Å–∫—É—é –æ–ø—Ç–∏–º–∏–∑–∞—Ü–∏—é
        self.automatic_optimization = False

    def forward(self, z):
        return self.generator(z)
    
    def training_step(self, batch, batch_idx):
        opt_d, opt_g = self.optimizers()  # –ü–æ–ª—É—á–∞–µ–º –æ–ø—Ç–∏–º–∏–∑–∞—Ç–æ—Ä—ã: –¥–∏—Å–∫—Ä–∏–º–∏–Ω–∞—Ç–æ—Ä–∞ –∏ –≥–µ–Ω–µ—Ä–∞—Ç–æ—Ä–∞
        real_images, _ = batch
        batch_size = real_images.size(0)
        device = real_images.device
        
        real_labels = torch.full((batch_size,), 1.0, device=device)
        fake_labels = torch.full((batch_size,), 0.0, device=device)
        
        # –û–±–Ω–æ–≤–ª–µ–Ω–∏–µ –¥–∏—Å–∫—Ä–∏–º–∏–Ω–∞—Ç–æ—Ä–∞
        opt_d.zero_grad()
        output_real = self.discriminator(real_images).view(-1)
        loss_real = self.criterion(output_real, real_labels)
        
        noise = torch.randn(batch_size, self.noise_dim, device=device)
        fake_images = self.generator(noise)
        output_fake = self.discriminator(fake_images.detach()).view(-1)
        loss_fake = self.criterion(output_fake, fake_labels)
        loss_D = loss_real + loss_fake
        
        self.manual_backward(loss_D)
        opt_d.step()
        
        # –û–±–Ω–æ–≤–ª–µ–Ω–∏–µ –≥–µ–Ω–µ—Ä–∞—Ç–æ—Ä–∞
        opt_g.zero_grad()
        noise = torch.randn(batch_size, self.noise_dim, device=device)
        fake_images = self.generator(noise)
        output = self.discriminator(fake_images).view(-1)
        loss_G = self.criterion(output, real_labels)
        
        self.manual_backward(loss_G)
        opt_g.step()
        
        self.log('loss_D', loss_D, prog_bar=True, on_epoch=True)
        self.log('loss_G', loss_G, prog_bar=True, on_epoch=True)
        Task.current_task().get_logger().report_scalar("Loss", "Discriminator", loss_D.item(), iteration=self.global_step)
        Task.current_task().get_logger().report_scalar("Loss", "Generator", loss_G.item(), iteration=self.global_step)
        return {"loss_D": loss_D, "loss_G": loss_G}

    def validation_step(self, batch, batch_idx):
        # –õ–æ–≥–∏—Ä–æ–≤–∞–Ω–∏–µ —Å–≥–µ–Ω–µ—Ä–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π, –µ—Å–ª–∏ —Ç–µ–∫—É—â–∞—è —ç–ø–æ—Ö–∞ –∫—Ä–∞—Ç–Ω–∞ debug_samples_epoch
        if self.current_epoch % self.hparams.debug_samples_epoch == 0:
            device = self.device
            fixed_noise = torch.randn(16, self.noise_dim, device=device)
            fake_images = self.generator(fixed_noise)
            grid = vutils.make_grid(fake_images, normalize=True)
            pil_img = transforms.ToPILImage()(grid.cpu())
            Task.current_task().get_logger().report_image("Validation Images", f"Epoch {self.current_epoch}", iteration=self.global_step, image=pil_img)
        self.log("val_dummy", 0, prog_bar=False)
    
    def configure_optimizers(self):
        optimizerD = Adam(self.discriminator.parameters(), lr=self.lr, betas=self.betas)
        optimizerG = Adam(self.generator.parameters(), lr=self.lr, betas=self.betas)
        return [optimizerD, optimizerG], []
```

```python
%%writefile gan_datamodule.py

import torch
from torch.utils.data import DataLoader
import torchvision.datasets as datasets
import torchvision.transforms as transforms
from lightning import LightningDataModule

class MNISTDataModule(LightningDataModule):
    def __init__(self, data_dir='./data', batch_size=64):
        super().__init__()
        self.data_dir = data_dir
        self.batch_size = batch_size
        self.transform = transforms.Compose([
            transforms.ToTensor(),
            transforms.Normalize((0.5,), (0.5,))  # –Ω–æ—Ä–º–∞–ª–∏–∑–∞—Ü–∏—è –¥–ª—è –¥–∏–∞–ø–∞–∑–æ–Ω–∞ [-1, 1]
        ])
    
    def prepare_data(self):
        # –°–∫–∞—á–∏–≤–∞–Ω–∏–µ –¥–∞–Ω–Ω—ã—Ö ‚Äì –≤—ã–∑—ã–≤–∞–µ—Ç—Å—è —Ç–æ–ª—å–∫–æ –æ–¥–∏–Ω —Ä–∞–∑
        datasets.MNIST(root=self.data_dir, train=True, download=True)
        datasets.MNIST(root=self.data_dir, train=False, download=True)
    
    def setup(self, stage=None):
        # –ü—Ä–∏ –æ–±—É—á–µ–Ω–∏–∏ —Å–æ–∑–¥–∞—ë–º train –∏ validation –Ω–∞–±–æ—Ä—ã
        if stage == "fit" or stage is None:
            self.mnist_train = datasets.MNIST(root=self.data_dir, train=True, transform=self.transform)
            # –ò—Å–ø–æ–ª—å–∑—É–µ–º —á–∞—Å—Ç—å —Ç–µ—Å—Ç–æ–≤–æ–≥–æ –Ω–∞–±–æ—Ä–∞ –∫–∞–∫ –≤–∞–ª–∏–¥–∞—Ü–∏—é
            self.mnist_val = datasets.MNIST(root=self.data_dir, train=False, transform=self.transform)
        if stage == "test":
            self.mnist_test = datasets.MNIST(root=self.data_dir, train=False, transform=self.transform)
    
    def train_dataloader(self):
        return DataLoader(self.mnist_train, batch_size=self.batch_size, shuffle=True, num_workers=4)
    
    def val_dataloader(self):
        # –ò—Å–ø–æ–ª—å–∑—É–µ–º –≤–∞–ª–∏–¥. –Ω–∞–±–æ—Ä, —Å–æ–∑–¥–∞–Ω–Ω—ã–π –≤ setup()
        return DataLoader(self.mnist_val, batch_size=self.batch_size, shuffle=False, num_workers=4)
```

```python
%%writefile train.py
#!/usr/bin/env python
import os
import argparse
from lightning import Trainer, seed_everything
from lightning.pytorch.callbacks import ModelCheckpoint
from gan_module import GAN
from gan_datamodule import MNISTDataModule
from clearml import Task
import torch

def check_clearml_credentials():
    required_vars = ["CLEARML_API_HOST", "CLEARML_WEB_HOST", "CLEARML_API_ACCESS_KEY", "CLEARML_API_SECRET_KEY"]
    missing = [var for var in required_vars if var not in os.environ]
    if missing:
        raise EnvironmentError(f"Missing ClearML credentials: {missing}")
    else:
        print("All ClearML credentials are present.")

def main():
    parser = argparse.ArgumentParser()
    parser.add_argument("--epoch", type=int, default=10, help="–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —ç–ø–æ—Ö –æ–±—É—á–µ–Ω–∏—è")
    parser.add_argument("--debug_samples_epoch", type=int, default=1, help="–ß–∞—Å—Ç–æ—Ç–∞ –ª–æ–≥–∏—Ä–æ–≤–∞–Ω–∏—è –æ—Ç–ª–∞–¥–æ—á–Ω—ã—Ö —Å—ç–º–ø–ª–æ–≤ (1 - –∫–∞–∂–¥—É—é —ç–ø–æ—Ö—É, 2 - –∫–∞–∂–¥—É—é –≤—Ç–æ—Ä—É—é –∏ —Ç.–¥.)")
    args = parser.parse_args()
    
    # –ü—Ä–æ–≤–µ—Ä–∫–∞ –Ω–∞–ª–∏—á–∏—è ClearML credentials
    check_clearml_credentials()
    
    # –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏—è ClearML –∑–∞–¥–∞—á–∏
    task = Task.init(project_name="GAN Experiment", task_name="GAN MNIST Lightning")
    
    # –õ–æ–≥–∏—Ä–æ–≤–∞–Ω–∏–µ –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ —á–µ—Ä–µ–∑ ClearML
    config = {
        "batch_size": 64,
        "lr": 0.0002,
        "num_epochs": args.epoch,
        "noise_dim": 100,
        "debug_samples_epoch": args.debug_samples_epoch
    }
    config = task.connect(config)
    
    # –£—Å—Ç–∞–Ω–æ–≤–∫–∞ seed –¥–ª—è –≤–æ—Å–ø—Ä–æ–∏–∑–≤–æ–¥–∏–º–æ—Å—Ç–∏
    seed_everything(42)
    
    # –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏—è –º–æ–¥–µ–ª–∏ GAN —Å –ø–µ—Ä–µ–¥–∞—á–µ–π debug_samples_epoch
    model = GAN(noise_dim=config["noise_dim"], lr=config["lr"], debug_samples_epoch=config["debug_samples_epoch"])
    
    # –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏—è DataModule –¥–ª—è MNIST
    dm = MNISTDataModule(batch_size=config["batch_size"])
    
    # –ù–∞—Å—Ç—Ä–æ–π–∫–∞ –∫–æ–ª–±—ç–∫–æ–≤: —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ —á–µ–∫–ø–æ–∏–Ω—Ç–æ–≤ –∏ –º–æ–Ω–∏—Ç–æ—Ä–∏–Ω–≥ learning rate
    checkpoint_callback = ModelCheckpoint(
        monitor="loss_G", mode="min", save_top_k=1, filename="gan-{epoch:02d}-{loss_G:.2f}"
    )
    
    trainer = Trainer(
        max_epochs=config["num_epochs"],
        accelerator="gpu" if torch.cuda.is_available() else "cpu",
        devices=1,
        callbacks=[checkpoint_callback],
        log_every_n_steps=50
    )
    
    # –ó–∞–ø—É—Å–∫ –æ–±—É—á–µ–Ω–∏—è
    trainer.fit(model, datamodule=dm)

    task.close()

if __name__ == "__main__":
    main()
```
